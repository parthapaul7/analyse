```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('analysis_results.csv')
```

```python
df.sort_values(by='Diff per Li Sites', ascending=False, inplace=True)
df
```

```python
# plot strucure as x asis and diff per Li sites as y axis
plt.figure(figsize=(10, 5))
# sns.set(style="whitegrid")
sns.barplot(x='Structure', y='Diff per Li Sites', data=df, palette="Blues_d")
plt.xticks(rotation=90)
plt.xlabel('Structure')
plt.ylabel('Diff per Li Sites')
plt.title('Diff per Li Sites for Different Structures')
plt.tight_layout()
plt.savefig('diff_per_li_sites.png', dpi=300)
plt.show()
```

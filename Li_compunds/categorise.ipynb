```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
df = pd.read_csv('analysis_results.csv')

df['Index'] = df.groupby('Structure').cumcount().astype(str)

df['Index'] = df['Structure'] + '_' + df['Index']
```

```python
df_sorted = df.sort_values(by='Diff per Li Sites', ascending=True, ignore_index=True)
df_sorted
```

```python
# plot strucure as x asis and diff per Li sites as y axis
plt.figure(figsize=(15, 5))
# sns.set(style="whitegrid")
sns.barplot(x='Index', y='Diff per Total Sites', data=df_sorted, palette="Blues_d",  errorbar=('ci', 0),  order=df_sorted['Index'])
plt.xticks(rotation=90)
plt.xlabel('Structure')
plt.ylabel('Diff per Li Sites')
plt.title('Diff per Li Sites for Different Structures')
plt.tight_layout()
plt.savefig('diff_per_li_sites.png', dpi=300)
plt.show()
```

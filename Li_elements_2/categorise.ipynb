```python
import pandas as pd

# 1. Load your data
analysis = pd.read_csv('analysis_results.csv')   # contains: Symbol, Structure, Total Sites, Li Sites, Counter Ion Sites, Result, ...
props    = pd.read_csv('all_props.csv')         # contains: Symbol, AtomicMass, AtomicRadius, Electronegativity, IonizationEnergy, ElectronAffinity, Polarizability, ...

# 2. Bin each structure into one of 30 groups by its “Result” value
analysis['Group'] = pd.cut(
    analysis['Result'],
    bins=30,
    labels=[f'Group {i}' for i in range(1, 31)]
)

# 3. Prepare the atomic‐property table
#    (one row per element, with its atomic descriptors)
atomic_cols = [
    'Symbol',
    'AtomicMass',
    'AtomicRadius',
    'Electronegativity',
    'IonizationEnergy',
    'ElectronAffinity',
    'Polarizability'
]
element_props = props[atomic_cols].drop_duplicates(subset='Symbol')

# 4. Merge the atomic properties into the per‐structure table
df = pd.merge(
    analysis,
    element_props,
    on='Symbol',
    how='left'
)

# 5. Define which columns to summarize per group
summary_cols = [
    'AtomicMass',
    'AtomicRadius',
    'Electronegativity',
    'IonizationEnergy',
    'ElectronAffinity',
    'Polarizability',
    'Total Sites',
    'Li Sites',
    'Counter Ion Sites'
]

# 6. Compute the mean and standard deviation for each property in each group
group_stats = (
    df
    .groupby('Group')[summary_cols]
    .agg(['mean', 'std'])
)

# 7. Flatten the MultiIndex column names
group_stats.columns = [
    f"{prop}_{stat}"
    for prop, stat in group_stats.columns
]

# 8. Save to CSV
group_stats.to_csv('group_summary.csv')

# Optional: print a preview
print(group_stats.head())
```

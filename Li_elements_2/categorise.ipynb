```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
df = pd.read_csv('analysis_results.csv')

df['Index'] = df.groupby('Structure').cumcount().astype(str)

df['Index'] = df['Structure'] + '_' + df['Index']
```

```python
df_sorted = df.sort_values(by='Diff per Li Sites', ascending=True, ignore_index=True)
df_sorted
```

```python
# plot strucure as x asis and diff per Li sites as y axis
plt.figure(figsize=(15, 5))
# sns.set(style="whitegrid")
sns.barplot(x='Index', y='Diff per Total Sites', data=df_sorted, palette="Blues_d",  errorbar=('ci', 0),  order=df_sorted['Index'])
plt.xticks(rotation=90)
plt.xlabel('Structure')
plt.ylabel('Diff per Li Sites')
plt.title('Diff per Li Sites for Different Structures')
plt.tight_layout()
plt.savefig('diff_per_li_sites.png', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data
# df = pd.read_csv('all_props.csv')

# Divide into 10 groups based on 'Result'
df['Group'] = pd.cut(df['Result'], bins=10, labels=[f'Group {i}' for i in range(1, 11)])

# Now 'Group' column has labels: Group 1, Group 2, ..., Group 10

# Plotting
plt.figure(figsize=(20, 10))
sns.barplot(x='Index', y='Diff per Total Sites', hue='Group', data=df, palette='tab10', errorbar=('ci', 0), dodge=False)
plt.xticks(rotation=90)
plt.xlabel('Structure (Index)')
plt.ylabel('Diff per Li Sites')
plt.title('Diff per Li Sites for Different Structures (Grouped by Result)')
plt.legend(title='Result Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('diff_per_li_sites_grouped.png', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data
# df = pd.read_csv('all_props.csv')

# Strip spaces from columns if needed
df.columns = df.columns.str.strip()

# Divide into 10 groups based on 'Result'
cats, edges = pd.cut(df['Result'], bins=15, retbins=True, labels=[f'Group {i}' for i in range(1,16)])
df['Group'] = cats

print("Bin edges used by pd.cut:")
for i in range(len(edges)-1):
    print(f"  Bin {i+1}: [{edges[i]:.4f}, {edges[i+1]:.4f})")

# Plot for each group separately
for group_label in df['Group'].unique():
    group_data = df[df['Group'] == group_label]
    
    plt.figure(figsize=(18, 6))
    sns.barplot(x='Index', y='Result', data=group_data, palette='viridis', errorbar=('ci', 0), order=group_data['Index'])
    
    plt.xticks(rotation=90)
    plt.xlabel('Structure (Index)')
    plt.ylabel('Diff per Li Sites')
    plt.title(f'Diff per Li Sites for Structures - {group_label}')
    plt.tight_layout()
    
    plt.show()
```

```python
## minimal errosrs


minimal = df[df['Result'] < 0.8]

# plot strucure as x asis and diff per Li sites as y axis
plt.figure(figsize=(15, 5))
# sns.set(style="whitegrid")
sns.barplot(x='Index', y='Diff per Total Sites', data=minimal, palette="Blues_d",  errorbar=('ci', 0),  order=minimal['Index'])
plt.xticks(rotation=35)
plt.xlabel('Structure')
plt.ylabel('Diff per Li Sites')
plt.title('Diff per Li Sites for Different Structures')
plt.tight_layout()
plt.savefig('diff_per_li_sites.png', dpi=300)
plt.show()
```

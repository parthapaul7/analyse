```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data
# df = pd.read_csv('all_props.csv')

# Strip spaces from columns if needed
df.columns = df.columns.str.strip()

# Divide into 20 groups based on 'Result'
cats, edges = pd.cut(df['Result'], bins=30, retbins=True, labels=[f'Group {i}' for i in range(1, 31)])
df['Group'] = cats

print("Bin edges used by pd.cut:")
for i in range(len(edges)-1):
    print(f"  Bin {i+1}: [{edges[i]:.4f}, {edges[i+1]:.4f})")

# List of properties you want to show
properties = [
    'AtomicMass',
    'AtomicRadius',
    'Electronegativity',
    'IonizationEnergy',
    'ElectronAffinity',
    'Polarizability',
    'Total Sites',
    'Li Sites',
    'Counter Ion Sites'
]

# Plot for each group separately
for group_label in df['Group'].unique():
    group_data = df[df['Group'] == group_label]
    
    # Merge group_data with props on 'Symbol'
    df_merged = group_data.merge(props, on='Symbol', how='left')
    
    # Calculate stats for each property
    text_lines = []
    for prop in properties:
        if prop in df_merged.columns:
            mean_val = df_merged[prop].mean()
            std_val = df_merged[prop].std()
            text_lines.append(f"{prop}: mean={mean_val:.2f}, std={std_val:.2f}")
        else:
            text_lines.append(f"{prop}: Not Found")  # in case column is missing

    # Join all lines into one text box
    textstr = '\n'.join(text_lines)

    # Start plotting
    plt.figure(figsize=(18, 6))
    sns.barplot(x='Index', y='Result', data=group_data, palette='viridis', errorbar=('ci', 0), order=group_data['Index'])

    plt.xticks(rotation=45)
    plt.xlabel('Structure (Index)')
    plt.ylabel('Energy Per Atom (Eh)')
    plt.title(f'Energy Per Atom for all Structures - {group_label}')
    
    # Add text box
    plt.gca().text(
        1.02, 0.8, textstr,
        transform=plt.gca().transAxes,
        fontsize=10,
        verticalalignment='top',
        bbox=dict(boxstyle='round,pad=0.5', facecolor='white', alpha=0.8)
    )

    plt.tight_layout()
    plt.savefig(f'imgs/{group_label}.png', dpi=300, bbox_inches='tight')
    plt.show()
```

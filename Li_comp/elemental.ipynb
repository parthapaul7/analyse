```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('all_props_new.csv')
```

```python
df['results_per_total'] = df['results_per_total'] * 27.2114 / 1000
```

```python
df
```

```python
df_more_than_2 = df[df['Electronegativity'] > 2]
df_less_than_2 = df[df['Electronegativity'] <= 2]

df_more_than_2
```

```python
element = 'C'
```

```python
grouped_df = df.groupby('Symbol')
oxy_group = pd.DataFrame()
for symbol, group in grouped_df:
    # print(f"Category: {symbol}")
    # print(group['Result'].max())
    # print(group['Result'].min())
    if symbol == element:
        # print(group['Result'])
        # print('---------------------')
        # print((group['Result'].max() - group['Result'].min())/group['Result'].mean() * 100)
        # print(group)
        oxy_group  = group
    # if(symbol == 'Al'):
    #     print(group)
    # print("\n")

oxy_group
```

```python
res = pd.read_csv('analysis_results.csv')
```

```python
res = res[res['Symbol'] == element]

res
```

```python
# oxy_group = oxy_group.sample(frac=0.5, random_state=42)
# sampled_oxy_group
```

```python
oxy_group 
```

```python
merged_df = pd.merge(res, oxy_group, left_index=True, right_index=True)
# merged_df.sort_values(by='Result',inplace=True, ascending=False)
merged_df['SpaceGroup']

merged_df['Counter Ion Sites']
```

```python
merged_df.rename(columns={"results_per_total": "Result_1"}, inplace=True)
merged_df['Total Sites_x']
```

```python
# merged_df.sort_values(,inplace=True, ascending=False)
```

```python
symbol = merged_df['Symbol_x'].values[0]
symbol
```

```python
full_li = merged_df['Li Sites']
merged_df['Anion Site'] = (merged_df['Counter Ion Sites'] / full_li).round(2)

merged_df.sort_values(by=['Anion Site','Result_1'],inplace=True, ascending=[True,False])
merged_df['Structure_SG'] = "Li-" + symbol+(merged_df['Counter Ion Sites'] / full_li).round(2).astype(str) 
```

```python
merged_df
```

```python
avg_electronegativity = merged_df['Electronegativity'].mean()
avg_polarizability = merged_df['Polarizability'].mean()
avg_result = merged_df['Result_1'].mean()

print("Average Electronegativity:", avg_electronegativity)
print("Average Polarizability:", avg_polarizability)
print("Average Result:", avg_result)
```

```python
merged_df.to_csv(f'chemical_space/Li_{symbol}.csv', index=False)
```

```python
# 1. Build a display label with subscripted CN
def make_disp_label(sg):
    # sg is like "Li-N0.33" or "K-N1.0"
    metal, cn_str = sg.split(f'-{element}')  # ["Li", "0.33"]
    cn_str = f"{float(cn_str):.2f}"  # Ensure two decimal places
    return rf"{cn_str}"

merged_df['DispLabel'] = merged_df['Structure_SG'].apply(make_disp_label)

# 2. Plot
x = np.arange(len(merged_df))
values = merged_df['Result_1']
labels = merged_df['DispLabel']

plt.figure(figsize=(16, 12))
plt.rcParams['font.family'] = 'Arial'
plt.bar(x, values, width=0.6, color='teal')

# 3. Use the LaTeX labels and set y-axis limit
plt.xticks(x, labels, rotation=45, ha='right', fontsize=38)
plt.yticks(fontsize=38)
plt.ylim(0, 0.35)  # Set y-axis limit to 0.5
plt.ylabel('Energy (eV)', fontsize=38)
plt.xlabel(rf'$\mathrm{{Li{symbol}_X}}$', fontsize=38)

plt.tight_layout()
plt.savefig(f'chemical_space/li_{symbol}_compounds.png', dpi=300, bbox_inches='tight')
plt.show()
```

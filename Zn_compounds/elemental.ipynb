```python
df = pd.read_csv('all_props_new.csv')
```

```python
element = 'Cl'
```

```python
grouped_df = df.groupby('Symbol')
oxy_group = pd.DataFrame()
for symbol, group in grouped_df:
    # print(f"Category: {symbol}")
    # print(group['Result'].max())
    # print(group['Result'].min())
    if symbol == element:
        # print(group['Result'])
        # print('---------------------')
        # print((group['Result'].max() - group['Result'].min())/group['Result'].mean() * 100)
        # print(group)
        oxy_group  = group
    # if(symbol == 'Al'):
    #     print(group)
    # print("\n")

oxy_group
```

```python
res = pd.read_csv('analysis_results.csv')
res
```

```python
res = res[res['Symbol'] == element]

res
```

```python
oxy_group
```

```python
merged_df = pd.merge(res, oxy_group, left_index=True, right_index=True)
# merged_df.sort_values(by='Result',inplace=True, ascending=False)
merged_df['SpaceGroup']

merged_df
```

```python
merged_df.rename(columns={"results_per_total": "Result_1"}, inplace=True)
merged_df['Total Sites_x']
```

```python
merged_df.sort_values(by='Result_1',inplace=True, ascending=False)
```

```python
symbol = merged_df['Symbol_x'].values[0]
symbol
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create combined x-axis label
merged_df['Structure_SG'] = merged_df['Reduced Formula'] + ' | ' + merged_df['SpaceGroup'].astype(str)

# Set data
x_labels = merged_df['Structure_SG']
values = merged_df['Result_1']

# Create figure
plt.figure(figsize=(10, 6))

# Create bar plot
x = np.arange(len(x_labels))  # numeric x-axis positions
plt.bar(x, values, width=0.6, label='Result_1')

# X-axis setup
plt.xticks(x, x_labels, rotation=60, ha='right')
plt.xlabel('Structure | SpaceGroup')
plt.ylabel('Value')
# plt.title('Grouped Bar Plot of Result_1 by Structure and SpaceGroup')
plt.legend(title='Metric')
plt.tight_layout()

# Save or show
plt.savefig(f'chemical_space/li_{symbol}_compounds.png', dpi=300, bbox_inches='tight')
plt.show()
```
